```python
# matplotlib.pyplotnumpy
import matplotlib.pyplot as plt
import numpy as np

# 
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 
x1 = [0, 1, 2, 3]
y1 = [1, 2, -5, 2]
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line Plot')

# sin
x2 = np.linspace(0, 10, 100)
y2 = 2 + 2 * np.sin(2 * x2)

axes[0, 1].plot(x2, y2, linewidth=2.0)
axes[0, 1].set_title('Sin Plot')
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11),
              ylim=(-1, 5), yticks=np.arange(-1, 6))

# 
x3 = np.random.normal(15, 5, 2000)
axes[0, 2].hist(x3, bins=30, alpha=0.7, color='green')
axes[0, 2].set_title('Histogram')

# 
np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))

axes[1, 0].scatter(x4, y4, c='red', marker='o')
axes[1, 0].set_title('Scatter Plot')

# 
plt.show()
```
