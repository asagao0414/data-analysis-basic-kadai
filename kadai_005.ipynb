```python
# matplotlib.pyplotnumpy
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)
plt.show()


# 
x = [0, 1, 2, 3]
y = [1, 2, -5, 2]
axes[0, 0]
plt.plot(x, y)


#sin
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0, 1]
plt.show()


# 
x = np.random .normal(15, 5, 2000)
axes[0, 2]
plt.hist(x)


# 
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

axes[1, 0]
plt.scatter(x, y)
plt.show()
```
